# Sales Dashboard Analysis Exploring the Superstore dataset to identify revenue trends, top products, and profit drivers.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('superstore.csv', encoding='latin1')

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Profit Margin'] = df['Profit'] / df['Sales']

df.dropna(inplace=True)
df.head()
```

Furniture seems to bring in high revenue but thin margins, Discounting it seems to bring it to a loss, especially with bulk orders. The pricing discount is probably to stay competetive for tables. Shipping costs may be eating into profits. 

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
region_sales.plot(kind='bar', title='Total Sales by Region', color='steelblue')
plt.ylabel('Sales ($)')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
